# Dataset 1: India_Exchange_Rate_Dataset.xls

```python
import pandas as pd
import matplotlib.pyplot as plt 
import warnings
%matplotlib inline
from statsmodels.tsa.filters.hp_filter import hpfilter
warnings.filterwarnings("ignore")

from pandas import datetime
from scipy import signal
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
df = pd.read_csv('/content/sample_data/India_Exchange_Rate_Dataset.csv', index_col=0,parse_dates=True)
df.head()
```

```python
EXINUS_cycle,EXINUS_trend = hpfilter(df['EXINUS'], lamb=1600)
EXINUS_trend.plot(figsize=(8,4)).autoscale(axis='x',tight=True)
```

1)	Implement detrending using differencing from scratch.

```python
series = pd.read_csv('/content/sample_data/India_Exchange_Rate_Dataset.csv', index_col=0,parse_dates=True)

X = series.values
diff = list()
for i in range(1, len(X)):
  value = X[i] - X[i - 1]
  diff.append(value)
plt.figure(figsize=(10,4))
plt.plot(diff)
plt.title('Detrending using Differencing from scratch', fontsize=16)
plt.xlabel('Year')
plt.ylabel('EXINUS exchange rate')
plt.show()
```

2)	Implement detrending using pandas differencing using diff() function.

```python
diff = df.EXINUS.diff()
plt.figure(figsize=(10,4))
plt.plot(diff)
plt.title('Detrending using Pandas Differencing', fontsize=16)
plt.xlabel('Year')
plt.ylabel('EXINUS exchange rate')
plt.show()
```

3)	Implement detrending using SciPy signal using signal.detrend submodule.

```python
detrended = signal.detrend(df.EXINUS.values)
plt.figure(figsize=(10,4))
plt.plot(detrended)
plt.xlabel('EXINUS')
plt.ylabel('Frequency')
plt.title('Detrending using Scipy Signal', fontsize=16)
plt.show()
```

4)	Implement detrending using HP filter using hp_filter submodule of statsmodels

```python
EXINUS_cycle,EXINUS_trend = hpfilter(df['EXINUS'], lamb=1600)
df['trend'] = EXINUS_trend
detrended = df.EXINUS - df['trend']
plt.figure(figsize=(10,4))
plt.plot(detrended)
plt.title('Detrending using HP Filter', fontsize=16)
plt.xlabel('Year')
plt.ylabel('EXINUS exchange rate')
plt.show()
```

# Dataset 3: monthly_milk.csv 

```python
df = pd.read_csv('https://raw.githubusercontent.com/ourcodingclub/CC-time-series/master/monthly_milk.csv')
df.head()
```

```python
df.rename(columns = {'milk_prod_per_cow_kg':'Milk_Production'}, inplace = True)
```

```python
Sales_cycle,Sales_trend = hpfilter(df['Milk_Production'], lamb=14400)
Sales_trend.plot(figsize=(12,4)).autoscale(axis='x',tight=True)
```

1)	Implement detrending using differencing from scratch.

```python
X = df.Milk_Production.values
diff = list()
for i in range(1, len(X)):
  value = int(X[i]) - int(X[i - 1])
  diff.append(value)
plt.figure(figsize=(15,4))
plt.plot(diff)
plt.title('Detrending using differencing from scratch', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Milk Production')
plt.show()
```

2)	Implement detrending using pandas differencing using diff() function.

```python
diff = df.Milk_Production.diff()
plt.figure(figsize=(15,4))
plt.plot(diff)
plt.title('Detrending using Pandas Differencing', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Milk Production')
plt.show()
```

3)	Implement detrending using SciPy signal using signal.detrend submodule.

```python
detrended = signal.detrend(df.Milk_Production.values)
plt.figure(figsize=(15,6))
plt.plot(detrended)
plt.xlabel('Milk Production')
plt.ylabel('Frequency')
plt.title('Detrending using Scipy Signal', fontsize=16)
plt.show()
```

4)	Implement detrending using HP filter using hp_filter submodule of statsmodels
